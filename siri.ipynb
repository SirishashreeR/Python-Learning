```python
card=input("did you insert the card? (yes/no): ")
print(card)
if card == "yes":
    print("you can proceed")
    pin =input("enter your pin:")
    if pin == "1234":
        print("correct pin")
        account = input("account is active? (yes/no): ")
        if account == "yes":
            print("you can withdraw money now")
            account_balance = input("account has sufficient balane? (yes/no): ")
            if account_balance == "yes":
                print("please collect your cash")
                atm_card = input("atm has sufficient cash? (yes/no): ")
                if atm_card == "yes":
                    print("please take your atm card")
                else:
                    print("atm out of cash, please contact bank")
            else:
                print("insufficient balance")
        else:
            print("account is inactive, please contact bank")
    else:
        print("incorrect pin")
else:
    print("please insert your card")
```
