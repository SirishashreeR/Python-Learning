```python
def prime_check(num):
    if num <= 1:
        return "not a prime number"
    
    for i in range(2, num):
        if num % i == 0:
            return "not a prime number"
    
    return "prime number"
```

```python
prime_check(10)
```

```python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)
```

```python
print(factorial(3))
```
